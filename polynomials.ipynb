# Legendre Polynomials

#### Legendre:

```python
matrix_size=15
t = np.linspace(-1, 1, 1000)
l = kl.integ.legendre(t, 30)
fig, ax = plt.subplots()
ax.plot(t, l[1], label=f"1-st Legendre")
ax.plot(t, l[2], label=f"2-nd Legendre")
ax.plot(t, l[3], label=f"3-rd Legendre")
for i in range(4,7):
    ax.plot(t, l[i], label=f"{i}-th Legendre")
ax.legend(loc='lower right')
ax.set_xlabel("x")
ax.set_ylabel("$\pi_i(x)$")
plt.show()
```

#### Normal Legendre:

```python
matrix_size=11
t = np.linspace(-1, 1, 1000)
l = kl.integ.legendre_normal(t, 50)
plt.plot(t, l[matrix_size])
plt.show()

fig, ax = plt.subplots()
ax.plot(t, l[1], label=f"1-st Legendre")
ax.plot(t, l[2], label=f"2-nd Legendre")
ax.plot(t, l[3], label=f"3-rd Legendre")
for i in range(4,7):
    ax.plot(t, l[i], label=f"{i}-th Legendre")
ax.legend(loc='lower right')
ax.set_xlabel("x")
ax.set_ylabel(r"$\tilde{\pi}_i(x)$")
plt.show()
```

#### Legendre over interval (a,b):

```python
matrix_size=15
a = 0
b = 10
t = np.linspace(a, b, 1000)
print(t.shape)
l = kl.integ.legendre_int(t, 50, a, b)
plt.plot(t, l[matrix_size])
```

#### Normal Legendre over interval (a,b):

```python
matrix_size=11
a = 3
b = 15
t = np.linspace(a, b, 1000)
print(t.shape)
l = kl.integ.legendre_normal_int(t, 100, a, b)
plt.plot(t, l[matrix_size])
plt.show()

fig, ax = plt.subplots()
ax.plot(t, l[1], label=f"1-st Legendre")
ax.plot(t, l[2], label=f"2-nd Legendre")
ax.plot(t, l[3], label=f"3-rd Legendre")
for i in range(4,7):
    ax.plot(t, l[i], label=f"{i}-th Legendre")
ax.legend(loc='lower right')
ax.set_xlabel("x")
ax.set_ylabel(r"$\tilde{\pi}_i(x)$")
plt.show()
```

#### 2D polynomials:

```python
nx, ny = (120, 100)
a = -1
b = 1
c = -1
d = 1
xx = np.linspace(a, b, nx)
yy = np.linspace(c, d, ny)
X, Y = np.meshgrid(xx, yy)
l = kl.integ.generate2DLegendre(X, Y, 500)
print(Y.shape)
print(X.shape)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, l[2])
plt.show()
```
