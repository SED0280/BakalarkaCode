# Legendre Polynomials

#### Legendre:
Generating first 6 monic Legendre polynomials

```python
matrix_size=15
t = np.linspace(-1, 1, 1000)
l = kl.integ.legendre(t, 30)
fig, ax = plt.subplots()
ax.plot(t, l[1], label=f"1-st Legendre")
ax.plot(t, l[2], label=f"2-nd Legendre")
ax.plot(t, l[3], label=f"3-rd Legendre")
for i in range(4,7):
    ax.plot(t, l[i], label=f"{i}-th Legendre")
ax.legend(loc='lower right', ncol=2)
ax.set_xlabel("$x$")
ax.set_ylabel("$\pi_i(x)$")
fig.savefig("Legendre.pdf", bbox_inches='tight')
plt.show()
```

#### Normal Legendre:

The first 6 normalized monic Legendre polynomials

```python
matrix_size=11
t = np.linspace(-1, 1, 1000)
l = kl.integ.legendre_normal(t, 50)
# plt.plot(t, l[matrix_size])
# plt.show()

fig, ax = plt.subplots()
ax.plot(t, l[1], label=f"1-st Legendre")
ax.plot(t, l[2], label=f"2-nd Legendre")
ax.plot(t, l[3], label=f"3-rd Legendre")
for i in range(4,7):
    ax.plot(t, l[i], label=f"{i}-th Legendre")
ax.legend(loc='lower right', ncol=3)
ax.set_xlabel("$x$")
ax.set_ylabel(r"$\tilde{\pi}_i(x)$")
fig.savefig("LegendreNorm.pdf", bbox_inches='tight')   
plt.show()
```
